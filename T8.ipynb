```python
import pandas as pd

#1
dataset = pd.read_csv('sampah.csv') # Membaca/import datagit add README.mdset yang original 
dataset.dropna(inplace=True) # Menghilangkan data yang tidak memiliki nilai angka atau akan berhasil nan
data_sampah = dataset[['nama_kabupaten_kota', 'jumlah_produksi_sampah', 'tahun']] # Mengambil kolom data hanya yang diperlukan
data_sampah.columns = ["Nama Kabupaten/Kota", "Jumlah Produksi Sampah (Ton)", "Tahun Pencatatan"] # Merubah nama/head kolom menjadi yang diinginkan
data_sampah.to_csv("data_sampah.csv", index=False) #hasil output file no 1
data_sampah.to_csv("data_sampah.xlsx", index=False)


#2
total = 0 # total diisikan nilai 0 untuk awalan 
tahun_dipilih = 2015 #Tahun ini bisa diganti sesuai yang kita inginkan

for _, baris in data_sampah.iterrows(): # Perulangan pada setiap baris di dataframe
    tahun = baris["Tahun Pencatatan"]
    if tahun == tahun_dipilih: # Apabila kondisi tahun sama maka jumlah produksi sampah perbaris akan dimasukan dan dijumlahkan di variable total
        total += baris["Jumlah Produksi Sampah (Ton)"]

if total > 0: # Apabila total telah terisi oleh penjumlahan dari jumlah produksi sampah maka akan keluar print jumlah produksi sampah
    print(f"Jumlah Produksi sampah di tahun {tahun_dipilih} adalah {total} (Ton)\n")
    hasil = pd.DataFrame({ # Dari total selanjunya dibuatkan menjadi dataframe
        "Tahun": [tahun_dipilih],
        "Total Sampah (Ton)": [total]
    })

    hasil.to_csv(f"Total_Sampah_Pada_Tahun_{tahun_dipilih}.csv", index=False) # #hasil output file no 2
    hasil.to_excel(f"Total_Sampah_Pada_Tahun_{tahun_dipilih}.xlsx", index=False)
else:
    print(f"Data pada tahun {tahun_dipilih} tidak ditemukan\n") # Kondisi apabila tahun dari data jumlah produksi sampah tidak ditemukan


#3
total_tahunan = {}  # Dictionary untuk menyimpan total per tahun sampah

for _, baris in data_sampah.iterrows():
    tahun = baris["Tahun Pencatatan"]
    if tahun in total_tahunan:
        total_tahunan[tahun] += baris["Jumlah Produksi Sampah (Ton)"]
    else:
        total_tahunan[tahun] = baris["Jumlah Produksi Sampah (Ton)"]

# Menampilkan total produksi sampah per tahun
print("Produksi Sampah Per  tahun:")
for tahun, total in total_tahunan.items():
    print(f"Total Produksi Sampah Pada Tahun {tahun}: {total} (Ton)")

print()#spasi

hasil_tahunan = pd.DataFrame(list(total_tahunan.items()), columns=[
    "Tahun",
    "Total Pertahun"
])
hasil_tahunan.to_csv("Total_Produksi_Sampah_Per_Tahun.csv", index=False) #hasil output file no 3
hasil_tahunan.to_excel("Total_Produksi_Sampah_Per_Tahun.xlsx", index=False)


#4
total_per_kab_kot = {}
per_kab_kot_df = []

for _, baris in data_sampah.iterrows():
    kab_kot = baris["Nama Kabupaten/Kota"]
    tahun = baris["Tahun Pencatatan"]
    
    key = (kab_kot, tahun)
    
    if key in total_per_kab_kot:
        total_per_kab_kot[key] += baris["Jumlah Produksi Sampah (Ton)"]
    else:
        total_per_kab_kot[key] = baris["Jumlah Produksi Sampah (Ton)"]

# Menampilkan total produksi sampah per kabupaten/kota dan per tahun
print("Produksi Sampah Per Kabupaten/Kota Per Tahun:")
for (kab_kot, tahun), total in total_per_kab_kot.items():
    print(f"Total Produksi Sampah Pada Kabupaten/Kota {kab_kot} Tahun {tahun}: {total} (Ton)")
    per_kab_kot_df.append([kab_kot, tahun, total])

per_kab_kot_df = pd.DataFrame(per_kab_kot_df, columns=[
    "Nama Kabupaten/Kota",
    "Tahun",
    "Total Pertahun"
])

per_kab_kot_df.to_csv("Total_Produksi_Per_Kabupaten_Kota_Per_Tahun.csv", index=False) #hasil output file no 4
per_kab_kot_df.to_excel("Total_Produksi_Per_Kabupaten_Kota_Per_Tahun.xlsx", index=False)


all_data = pd.concat([data_sampah, hasil, hasil_tahunan, per_kab_kot_df], ignore_index=True)

# Menyimpan hasil ke file csv dan xlsx
all_data.to_csv("Sampah_Prov_Jawa_Barat.csv", index=False)
all_data.to_excel("Sampah_Prov_Jawa_Barat.xlsx", index=False)
```
